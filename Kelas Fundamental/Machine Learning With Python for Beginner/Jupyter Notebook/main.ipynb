```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

# Load dataset
dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

# Imputation missing value
dataset.fillna(dataset.mean(), inplace=True)

# Proses Scalling Data
scaler = MinMaxScaler()
# Seleksi Features mana yang akan di Scaling datanya
data_column = (dataset.dtypes[dataset.dtypes == np.float64]).index
scalling_column = list(data_column)
scalling_column.remove('SpecialDay')
# Applikasi proses scaling data features yang telah dipilih
dataset[scalling_column] = scaler.fit_transform(dataset[scalling_column])

# Proses Konversi String ke Numeric
LE = LabelEncoder()
# Seleksi features mana yang akan di konversi dari string ke numeric
data_column_object = (dataset.dtypes[dataset.dtypes == "object"]).index
encoder_column = list(data_column_object)
display(len(encoder_column))
# Aplikasi proses encoder padan data features yang telah di pilih
# Pada Kolom #1
dataset[encoder_column[0]] = LE.fit_transform(dataset[encoder_column[0]])
display(LE.classes_)
display(np.sort(dataset[encoder_column[0]].unique()))
# Pada Kolom #2
dataset[encoder_column[1]] = LE.fit_transform(dataset[encoder_column[1]])
display(LE.classes_)
display(np.sort(dataset[encoder_column[1]].unique()))

# Features & Label
# Proses Memisahkan Label dan Features
# Removing the target column Revenue from dataset and assigning to X
X = dataset.drop(["Revenue"], axis=1)
# Assiging the target column Revenue to y
y = dataset["Revenue"]
# Checking the shapes
print("Shape of X = %d Baris %d Kolom " % X.shape)
print("Shape of y = %d Baris " % y.shape)
```

```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# Load dataset 
dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

# Imputasi missing value dengan nilai mean masing-masing kolom bersangkutan
dataset.fillna(dataset.mean())

# Merubah interval atau skala pada masing-masing features yang bersangkutan
# Definisikan fungsi MinMaxScaler
scaler = MinMaxScaler()
# Seleksi features mana saja yang akan di ubah interval nilai min - maxnya
data_column_float64 = (dataset.dtypes[dataset.dtypes == np.float64]).index
scalling_column = list(data_column_float64)
scalling_column.remove("SpecialDay")
# Aplikasi scalling data pada features yang telah diseleksi
dataset[scalling_column] = scaler.fit_transform(dataset[scalling_column])

# Merubah tipe data string ke numerik (Encoder)
# Definisikan fungsi LabelEncoder()
LE = LabelEncoder()
# Seleksi features mana saja yang akan di ubah dari tipe data string ke numerik
data_column_encoder = (dataset.dtypes[dataset.dtypes == "object"]).index
encoder_column = list(data_column_encoder)
print(len(encoder_column))
# Aplikasi encoder data pada feature yang telah diseleksi
dataset[encoder_column[0]] = LE.fit_transform(dataset[encoder_column[0]])
display(LE.classes_)
display(np.sort(dataset[encoder_column[0]].unique()))
dataset[encoder_column[1]] = LE.fit_transform(dataset[encoder_column[1]])
display(LE.classes_)
display(np.sort(dataset[encoder_column[1]].unique()))

# Features & Label
# Proses breakdown Features dan Label
# Features deklarasikan sebagai Variabel X
name_label = "Revenue"
X = dataset.drop([name_label], axis=1)
# Label deklarasikan sebagai Variabel y
y = dataset[name_label]
# Mengecek dimensi dari Features
display("Dimensi Features = %d Baris dan %d Kolom " %X.shape)
# Mengecek dimensi dari Label
display("Dimensi Label = %d Baris " %y.shape)

# Training dan Test Dataset
# Proses Breakdown masing-masing Fetaures dan Label ke Training dataset dan Test dataset
# Splitting the X dan y
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Checking the shapes
display("Dimensi X_train = %d Baris dan %d Kolom " %X_train.shape)
display("Dimensi y_train = %d Baris" %y_train.shape)
display("Dimensi X_test = %d Baris dan %d Kolom" %X_test.shape)
display("Dimensi y_test = %d Baris" %y_test.shape)
```

## Training Model : Fit

```python
import numpy as np
import pandas as pd
from IPython.display import display
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

# Load dataset
dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/online_raw.csv")

# Informasi dataset
display(dataset.columns.to_list())
display("Informasi Dataset : ")
display(dataset.info())
dataset_dtypes = dataset.dtypes.value_counts().index.to_list()
display(dataset_dtypes)
dataset_numeric = dataset.select_dtypes(include=dataset_dtypes[0:2])
display(dataset_numeric)
dataset_object = dataset.select_dtypes(include=dataset_dtypes[2])
display(dataset_object)
dataset_boolean = dataset.select_dtypes(include=dataset_dtypes[3])
display(dataset_boolean)

# Mengecek missing value
print("Data Missing Value Sebelum Proses Imputasi")
display(dataset.isnull().sum())

# Imputasi missing value dengan nilai rata-rata masing-masing kolom
for features_numeric in dataset_numeric.columns.to_list():
    dataset[features_numeric].fillna(dataset[features_numeric].mean(), inplace=True)

# Mengecek missing value
print("Data Missing Value Setelah Proses Imputasi")
display(dataset.isnull().sum())

# Mengecek interval nilai masing-masing features numeric
display(dataset.describe().T[["min","max"]])

# Merubah intervali nilai masing-masing features numeric menjadi 0 - 1
# Definisikan fungsi yang digunakan untuk mengubah interval masing-masing features numeric menjadi 0 - 1
scaler = MinMaxScaler()
# Seleksi features yang akan diubah skala nilainya
data_column_float64 = (dataset.dtypes[dataset.dtypes == "float64"]).index.to_list()
data_column_float64.remove("SpecialDay")
display(data_column_float64)
# Aplikasi fungsi scaler pada features yang telah diseleksi
dataset[data_column_float64] = scaler.fit_transform(dataset[data_column_float64])

# Merubah tipe data fetaures string menjadi numeric atau proses encoder
# Definisikan fungsi yang digunakan untuk mengubah tipe data features string menjadi numeric
LE = LabelEncoder()
# Seleksi features yang akan diubah menajdi tipe data numeric
data_column_object = (dataset.dtypes[dataset.dtypes == "object"]).index.to_list()
# Aplikasi fungsi LE pada features yang telah diseleksi
display(data_column_object)
dataset[data_column_object[0]] = LE.fit_transform(dataset[data_column_object[0]])
display(LE.classes_)
display(np.sort(dataset[data_column_object[0]].unique()))

dataset[data_column_object[1]] = LE.fit_transform(dataset[data_column_object[1]])
display(LE.classes_)
display(np.sort(dataset[data_column_object[1]].unique()))

# Breakdown Features dari Label
name_label = "Revenue"
X=dataset.drop(name_label, axis=1)
y=dataset[name_label]
# Mengecek dimensi Features dan Label
display("Dimensi Features = %d baris dan %d kolom" %X.shape) 
display("Dimesni Label = %d baris"%y.shape)

# Breakdown data features & label menjadi data training & test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
# Mengecek dimesin masing-masing data training & test
print("Dimensi X Training = %d Baris dan %d Kolom" %X_train.shape)
print("Dimensi y Training = %d Baris" %y_train.shape)
print("Dimensi X Test = %d Baris dan %d Kolom" %X_test.shape)
print("Dimensi y Test = %d Baris" %y_test.shape)

# Training Model 
# Definisikan model
model_DTC = DecisionTreeClassifier()
# Melatih model dengan data Training
model_DTC = model_DTC.fit(X_train, y_train)
```
