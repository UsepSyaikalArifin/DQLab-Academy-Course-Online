## Unsupervised Learning -> Measuring Cluster Criteria

```python
# Load library atau modul
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display
from sklearn.cluster import KMeans

# Load dataset
dataset = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/pythonTutorial/mall_customers.csv")
display(dataset.head(5))

# Informasi dataset
print("Dimensi dataset = %d Baris dan %d Kolom "%dataset.shape)
tipe_dataset = dataset.dtypes.to_frame("Tipe Data")
missing_value_dataset = dataset.isnull().sum().to_frame("Missing Value")
statistik_dasar_dataset = dataset.describe().T[["min","max","mean","25%","50%","75%","std"]]
resume_dataset = pd.concat([tipe_dataset, missing_value_dataset, statistik_dasar_dataset], axis = 1)
display(resume_dataset.sort_values(by="Tipe Data"))

# Memilih Features
kolom_features = ["annual_income","spending_score"]
X = dataset[kolom_features]

# Pemodelan
inertia = []
for i in range(1,10):
    model_KMeans = KMeans(n_clusters = i, random_state = 24)
    model_KMeans.fit(X)
    inertia_value = model_KMeans.inertia_
    inertia.append(inertia_value)

# Plot
plt.plot(range(1, 10), inertia, "bo")
plt.title("The Elbow Method - Inertia Plot", fontsize = 25)
plt.xlabel("No. of clusters", fontsize=15)
plt.ylabel("Inertia", fontsize=15)
plt.show()

# Pemodelan dengan KMeans
model_KMeans = KMeans(n_clusters=5, random_state=24)
model_KMeans = model_KMeans.fit(X)
y = model_KMeans.fit_predict(X)
df_y = pd.DataFrame(y, columns=["Label"])

# New Dataframe
new_dataset = pd.concat([dataset, df_y], axis=1)
display(new_dataset)

# Mengubah Dataframe X -> Array X
X = X.values
xs = X[:,0]
ys = X[:,1]

# Mendapatkan centorids model_KMeans
centroids = model_KMeans.cluster_centers_
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]

# Plotting data
plt.scatter(xs, ys, c=y, alpha=0.9)
plt.scatter(centroids_x, centroids_y, marker="D", s=100)
plt.title("KMeans CLustering", fontsize = 25)
plt.xlabel("Annual Income", fontsize = 15)
plt.ylabel("Spending Score", fontsize = 15)
plt.show()
```
