```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/data_retail.csv", sep=";")
display(df.head())

df["First_Transaction"] = pd.to_datetime(df['First_Transaction'], unit="ms")
df["Last_Transaction"] = pd.to_datetime(df['Last_Transaction'], unit="ms")
display(df.head())

display(max(df["Last_Transaction"]))

df.loc[df["Last_Transaction"] <= "2018-08-01", "is_churn"] = True
df.loc[df["Last_Transaction"] > "2018-08-01", "is_churn"] = False

display(df.head())

df = df.drop(["no","Row_Num"], axis = 1)
display(df.head())

df["Year_First_Transaction"] = df["First_Transaction"].dt.year
df["Year_Last_Transaction"] = df["Last_Transaction"].dt.year

display(df.head())

df_year_customerID = df.groupby(["Year_First_Transaction"])["Customer_ID"].count()
df_year_customerID.plot(x="Year_First_Transaction", y="Customer_ID", kind="bar", title="Graph of Customer Acquisition")
plt.xlabel("Year_First_Transaction")
plt.ylabel("Num_of_Customer")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/data_retail.csv", sep=";")
df["First_Transaction"] = pd.to_datetime(df["First_Transaction"], unit="ms")
df["Last_Transaction"] = pd.to_datetime(df["Last_Transaction"], unit="ms")
display(df.head())

print(max(df["Last_Transaction"]))
df.loc[df["First_Transaction"] <= "2018-08-01", "is_churn"] = True
df.loc[df["Last_Transaction"] > "2018-08-01", "is_churn"] = False
display(df.head())

df["First_Year_Transaction"] = df["First_Transaction"].dt.year
df["Last_Year_Transaction"] = df["Last_Transaction"].dt.year
display(df.head())

df_year_customerID = df.groupby(["First_Year_Transaction"])["Customer_ID"].count()
df_year_customerID.plot(x="First_Transaction", y="Customer_ID", kind="bar", title="Graph of Transaction Customer")
plt.xlabel("Firs_Year_Transaction")
plt.ylabel("Num_of_Customer")
plt.show()
df_year_countTransaction = df.groupby(["First_Year_Transaction"])["Count_Transaction"].sum()
df_year_countTransaction.plot(x="First_Transaction", y="Count_Transaction", kind="bar", title="Graph of Transaction Customer")
plt.xlabel("Firs_Year_Transaction")
plt.ylabel("Num_of_Transaction")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display

df = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/data_retail.csv", sep=";")

df["First_Transaction"] = pd.to_datetime(df["First_Transaction"], unit="ms")
df["Last_Transaction"] = pd.to_datetime(df["Last_Transaction"], unit="ms")
df["Year_First_Transaction"] = df["First_Transaction"].dt.year
df["Year_Last_Transaction"] = df["Last_Transaction"].dt.year
print(max(df["Last_Transaction"]))
df.loc[df["Last_Transaction"] <= "2018-08-01", "is_churn"] = True
df.loc[df["Last_Transaction"] > "2018-08-01", "is_churn"] = False
df = df.drop(["no","Row_Num"], axis=1)
display(df.head())
df_year_customerID = df.groupby(["Year_First_Transaction"])["Customer_ID"].count()
df_year_customerID.plot(x="Year_First_Transaction", y="Customer_ID", kind="bar", title="Graph of Transaction Customer")
plt.xlabel("First Year Transaction")
plt.ylabel("Customer ID")
plt.show()
df_year_countTransaction = df.groupby(["Year_First_Transaction"])["Count_Transaction"].sum()
df_year_countTransaction.plot(x="Year_First_Transaction", y="Count_Transaction", kind="bar", title="Graph of Transaction Count")
plt.xlabel("First Year Transaction")
plt.ylabel("Count Transaction")
plt.show()
display(df_year_customerID)
df_mean_product_of_year = df.groupby(["Product","Year_First_Transaction"]).mean().reset_index()
sns.pointplot(data=df_mean_product_of_year, x="Year_First_Transaction", y="Average_Transaction_Amount", hue="Product")
plt.show()
```
